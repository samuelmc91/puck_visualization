```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image

ROOT = os.getcwd()

def parquet_images():
    my_images = []
    for dirpath, dirnames, files in os.walk(ROOT):
        for my_file in files:
            filename, exten = os.path.splitext(my_file)
            if exten == '.jpg':
                my_images.append(os.path.join(dirpath, my_file))

    image_array = []
    for image in my_images:
        img = cv2.imread(image,0)
        img_out = img.tolist()
        image_array.append([[img_out], image, list(img.shape)])

    df = pd.DataFrame(image_array, columns=['Image', 'Path', 'Shape'])
    df.to_parquet('images.parquet', compression='gzip')

def restore_images():
    df = pd.read_parquet('images.parquet')
    for index, row in df.iterrows():
        img = Image.fromarray(np.array(row['Image'][1][0], dtype='uint8').reshape(tuple(row['Shape'][1])))
        img.save(row['Path'])
```
